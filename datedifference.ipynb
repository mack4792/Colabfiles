```python

def datdiff (dataframe):
    #First we will check if there are any null values, if yes, we will then remove those rows
    if dataframe.isnull().sum().sum()>0:
        dataframe.dropna()
    # Let us try converting the object type columns into datetime columns
    for i in dataframe.select_dtypes(include = 'object'):
        try:
            df[i] = pd.to_datetime(dataframe[i])
        except:
            df[i] = df[i]
    # Creating the list of names of new date difference column
    newnames = []
    col = dataframe.select_dtypes(include='datetime64[ns]').columns
    for j in range(len(col)):
        if j <len(col)-1:
            so = col[j] +"-"+ col[j+1]
            newnames.append(so)
            
        else:
            so = col[0] + "-" + col[j]
            newnames.append(so)
    # Now let us create a new difference dataframe
    new = pd.DataFrame()
    temp = pd.DataFrame()
    temp1 = dataframe.select_dtypes(include='datetime64[ns]')
    # Now let us difference the columns with dates
    for k,l in zip(range(len(temp1.columns)),col):
        if k+1 <= len(temp1.columns) - 1: #Because the indexing in python starts from 0
            temp = temp1.iloc[:,k]- temp1.iloc[:,k+1]
            temp = temp.dt.days
            
            new = pd.concat([new,temp], axis=1)
        else:
            temp = (temp1.iloc[:,0]- temp1.iloc[:,k]).dt.days
            
            new = pd.concat([new,temp], axis=1)
    new.columns = newnames
    return new
            
    

    
            
```
